```python
import pandas as pd
import numpy as np
```

```python
name = ['爸爸',
        '爸',
        '妈妈',
        '妈',
        '舅舅',
        '舅',
        '姑姑', 
        '姑', 
        '奶奶',
        '爷爷',
        '大爷',
        '二舅',
        '姨夫',
        '姨爹',
        '同事',
        '工友',
        '朋友',
        '同学', 
        '同窗',
        '前女友', 
        '女朋友', 
        '前男友', 
        '男朋友', 
        '房东', 
        '上司', 
        '老板', 
        '亲戚', 
        '邻居',  
        ]
prefix = ['我是她', '我是 她', '我 是 她', '我 是她', '我是 她的', '我 是 她的',
          '我是他', '我是 他', '我 是 他', '我 是他', '我是 他的', '我 是 他的',
          '他是 我', '他 是 我', '他是 我的', '他 是 我的', 
          '她是 我', '她 是 我', '她是 我的', '她 是 我的']
sens = []

ext_0 = ['我 认识','我 知道','恩','嗯','嗯哼','对呀','对','了解','听说 过',
         '听说过','晓得','我 认识',
         '他 就 住 我 旁边', '他 就 在 我 旁边', '他 就 住 我 楼下', '我 明早 就能 见到 他','我 有 听说过 他',
         '我和他 是 同事', '我和他 是 朋友', '他 好像 最近 电话 停机了', '他 离开 这个 城市 了',
        '她 就 住 我 旁边', '她 就 在 我 旁边', '他 就 住 我 楼下', '我 明早 就能 见到 她','我 有 听说过 她',
         '我和她 是 同事', '我和她 是 朋友', '她 好像 最近 电话 停机了', '她 离开 这个 城市 了', '我 现在 没 和她 在一起','我 现在 没 和她 在一起', '我 现在 没有 和她 在一起']


for p in prefix:
    for postfix in name:
        sens.append(p + ' ' + postfix)
sens.extend(ext_0)
label = np.zeros(len(sens))

df_0 = pd.DataFrame({'split_text':sens, 'label':label})
df_0['label'] = df_0['label'].astype('int')
```
