```python
import pandas as pd
import numpy as np
import re


mapping = {1:['IDClassifier',1],
 2:['IDClassifier',0],
 3:['IDClassifier',2],
 4:['IfKnowDebtor',0],
 5:['IfKnowDebtor',1], 
 6:['IfKnowDebtor',2],
 7:['WillingToPay',2], 
 8:['WillingToPay',0], 
 9:['WillingToPay',1], 
 10:['WillingToPay',3], 
 11:['SetDueDay',0], 
 12:['SetDueDay',1],
 13:['SetDueDay',2], 
 14:['CutDebt',0], 
 15:['CutDebt',1], 
 16:['CutDebt',2],
 17:['Installment',0],
 18:['Installment',1],
 19:['Installment',2]}
```

```python
df = pd.read_csv('../data/data_0612.csv',encoding='utf-8')
df = df.drop(df.index[[12]])
```

```python
result = {}
columns = df.columns.values

for i in range(1,20):
    expression = r'(?u)第{}题'.format(i)
    text = []
    delete = []
    for column in columns:
        if len(re.findall(expression,column)) > 0:
            delete.append(False)
            text.extend(list(df[column].values))
        else:
            delete.append(True)
    label = [mapping[i][1] for a in range(len(text))]
    columns = columns[delete]
    cur = {mapping[i][0]:{'split_text':text, 'label':label}}
    result[i] = cur
```

# Combine

```python
com_dict = {}
for each in result:
    key = list(result[each].keys())[0]
    if com_dict.get(key) is None:
        com_dict[key] = []
    com_dict[key].append(pd.DataFrame(result[each][key]))
result = {}
for each in com_dict:
    result[each] = pd.concat(com_dict[each])
```
