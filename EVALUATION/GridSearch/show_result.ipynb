```python
import pandas as pd
import sys,os
```

```python
__file__ = '../GridSearch/'
param_path = os.path.join(os.path.dirname(__file__), 'parameter/parameter_{}.csv')

report_path = os.path.join(os.path.dirname(__file__), 'report/report_{}.csv')
```

```python
model_list = ['IDClassifier', 'CutDebt', 'WillingToPay','IfKnowDebtor','Installment','ConfirmLoan']
```

```python
report_dict = {}
param_dict = {}
for model in model_list:
    param_path_evl = param_path.format(model)
    report_path_evl = report_path.format(model)
    param_dict[model] = pd.read_csv(param_path_evl)
    report_dict[model] = pd.read_csv(report_path_evl)
    
```

```python
id_r = report_dict['Installment']
```

```python
id_r.sort_values('accuracy_01',ascending=False)
```
