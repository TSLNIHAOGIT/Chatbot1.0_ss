```python
mapping = strategy.set_index('category').label.drop_duplicates()
```

```python
folder = 102
df = pd.read_csv('../../MLModel/data/others/labels/{}/mock_up_data.csv'.format(folder))
```

```python
df = df.rename({'类别':'category','文本':'text'},axis=1)
df['label'] = df['category'].map(mapping)
df = df[['text','category','label',
         'IDClassifier','IfKnowDebtor',
         'ConfirmLoan','WillingToPay',
         'CutDebt','Installment']]
print(df.label.value_counts())
```

```python
for folder in range(102,111):
    print(folder)
    df = pd.read_csv('../../MLModel/data/others/labels/{}/mock_up_data.csv'.format(folder))
    df = df.rename({'类别':'category','文本':'text'},axis=1)
    df['label'] = df['category'].map(mapping)
    df = df[['text','category','label',
             'IDClassifier','IfKnowDebtor',
             'ConfirmLoan','WillingToPay',
             'CutDebt','Installment']]
    df = df.fillna(-1)
    print(df.label.value_counts())
    df.to_csv('../../MLModel/data/others/labels/{}/mock_up_data.csv'.format(folder),index=False,encoding='utf8')
    
```
